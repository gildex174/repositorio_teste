```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
data
```

```python
print(data.shape[0], data.shape[1])
```

```python
data['id']
```

```python
X = data.iloc[:,2:32]
y = data['diagnosis']
```

```python
y
```

```python
y_arr = y.to_numpy()

counts = {'Malignant':0, 'Benign':0}

for value in y_arr:
    if value=='B':
        counts['Benign']+=1
    else:
        counts['Malignant']+=1


labels = list(counts.keys())
values = list(counts.values())
colors = ['red', 'blue']

plt.bar(labels, values, color=colors)
plt.grid(axis='y', linestyle='dotted', linewidth=0.5, alpha=1)
plt.show()
```

```python
X
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(X_scaled)
```

```python
X_pca = pca.transform(X_scaled)
```

```python
principal_components = pd.DataFrame(X_pca, columns=['Componente Principal 1', 'Componente Principal 2'])
result = pd.concat([principal_components, y], axis=1)
print(result.head())
```

```python
print(result)
```

```python
X_reconstructed = pca.inverse_transform(X_pca)
```

```python
mse = mean_squared_error(X_scaled, X_reconstructed)
print(mse)
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1])

plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('GrÃ¡fico de DispersÃ£o dos Pontos Transformados')

plt.show()
```
